```python
#TOKENIZING
```

```python
from transformers import AutoTokenizer
```

```python
tokenizer=AutoTokenizer.from_pretrained("google/bert_uncased_L-2_H-128_A-2")
```

```python
encoded=tokenizer(train_dataset['sentence'])
```

```python
tokenizer.decode(tokenizer(train_dataset[0]['sentence'])['input_ids'])
```

```python
def encode(examples):
    return tokenizer(
        examples['sentence'],
        truncation=True,
        padding='max_length',
        max_length=512
    )
```

```python
train_dataset=train_dataset.map(encode,batched=True)
```

```python
train_dataset[0]
```

```python
#FORMATTING
import torch
```

```python
train_dataset.set_format(type="torch",columns=['input_ids','attention_mask','label'])
```

```python
train_dataset[0]
```

```python
#DATA LOADER 
data_loader=torch.utils.data.DataLoader(train_dataset,batch_size=32)
```
